## 1. Create a function which takes in a set of coordinates (lat and long) and another set, and calculates the distance between them.

```python
df.head()
```

```python
def distance_between_coordinates(lat_1, lon_1,lat_2, lon_2):
    lat_1= lat_1.astype(float) #convert to float
    long_1= long_1.astype(float)
    lat_2= lat_2.astype(float)
    long_2= long_2.astype(float)

    radius_of_earth = 6371.01

    d_lat = lat_2 - lat_1
    d_lon = lon_2 - lon_1

    a = (d_lat**2) + (d_lon**2)
    c = math.sqrt(a)

    distance = radius_of_earth * c

    return distance
```

## Research the usage of .at and .iat, include examples and explanations using the housing dataset and custom datasets you created

The .at and .iat methods in pandas are used to access individual elements within a DataFrame or Series based on their labels or integer positions, respectively. These methods are designed for fast and efficient single-item access. Here's how they work along with examples using the housing dataset and a custom dataset:

**.at Method: <br>
The .at method is used to access a single element in a DataFrame using row and column labels.** <br>
- df.at[row_label, column_label]


```python
price0 = df.at[0, 'price']
print("Price of house at index 0:", price0)
```

**.iat Method:<br>
The .iat method is used to access a single element in a DataFrame using integer positions for both rows and columns.**<br>
- df.iat[row_position, column_position]

```python
element01 = df.iat[0, 1]
print("Element at (0, 1):", element01)
```
